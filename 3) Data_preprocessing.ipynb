```python
#tweets = tweets.withColumn('word_count',f.size(f.split(f.col('text'),' ')))
#tweets.show()

df_tweets = df_tweets.withColumn('word_count',f.size(f.split(f.col('text'),' ')))
df_tweets.show()
```

```python
#Vader sentiment analysis
import nltk
import re
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
sid = SIA()
#tweets=tweets.toPandas()
#tweets['sentiments'] = tweets['text'].apply(lambda x: sid.polarity_scores(' '.join(re.findall(r'\w+',x.lower()))))
#tweets.head()

df_tweets=df_tweets.toPandas()
df_tweets['sentiments'] = df_tweets['text'].apply(lambda x: sid.polarity_scores(' '.join(re.findall(r'\w+',x.lower()))))
df_tweets.head()
```

```python
#pd.set_option('display.max_colwidth', None)
#tweets.head()

pd.set_option('display.max_colwidth', None)
df_tweets.head()
```

```python
#tweets['Positive Sentiment'] = tweets['sentiments'].apply(lambda x: x['pos']+1*(10**-6)) 
#tweets['Neutral Sentiment'] = tweets['sentiments'].apply(lambda x: x['neu']+1*(10**-6))
#tweets['Negative Sentiment'] = tweets['sentiments'].apply(lambda x: x['neg']+1*(10**-6))
#tweets['Overall Sentiment']=tweets['sentiments'].apply(lambda x: x['compound']+1*(10**-6))

#tweets.drop(columns=['sentiments'],inplace=True)
#tweets.head()


df_tweets['Positive Sentiment'] = df_tweets['sentiments'].apply(lambda x: x['pos']+1*(10**-6)) 
df_tweets['Neutral Sentiment'] = df_tweets['sentiments'].apply(lambda x: x['neu']+1*(10**-6))
df_tweets['Negative Sentiment'] = df_tweets['sentiments'].apply(lambda x: x['neg']+1*(10**-6))
df_tweets['Overall Sentiment']=df_tweets['sentiments'].apply(lambda x: x['compound']+1*(10**-6))

df_tweets.drop(columns=['sentiments'],inplace=True)
df_tweets.head()
```

```python
import numpy as np
df_tweets['Overall Sentiment Description'] = np.where(df_tweets['Overall Sentiment']>=0.5, 'Positive', 
    np.where(df_tweets['Overall Sentiment']<=-0.05, 'Negative', 'Neutral'))
df_tweets.head()
```
